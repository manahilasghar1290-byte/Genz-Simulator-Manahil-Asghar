```python
import time  # module to run the program smoothly and make it realistic

user = {}  # dict to store name and age

responses = {  # dict to store responses
    "hi": "Hey bestie! üëã",
    "hello": "Hey bestie! üëã",
    "how are you": "Just vibing, you? üòé",
    "bye": "Aww, see you soon üíñ",
}

state = {  # dict to store status of player
    "energy": 80,
    "mood": "neutral",
    "points": 0,
    "room_level": 1,
    "decor_items": [],
}

shop_items = {
    "poster": {"cost": 10, "desc": "Cute artistic poster ‚Äî +mood"},
    "plants": {"cost": 15, "desc": "Aesthetic plants ‚Äî chill vibes"},
    "lamp": {"cost": 20, "desc": "Cozy lamp ‚Äî better energy"},
    "beanbag": {"cost": 25, "desc": "Comfy beanbag ‚Äî relax mode"},
    "rgb lights": {"cost": 35, "desc": "RGB lights ‚Äî cool vibe"},
}

def GreetUser():
    print("‚ú® Welcome to GenZ Life Simulator ‚Äî Room Edition ‚ú®")
    name = input("What's your name? ").strip()
    while not name:
        name = input("Please type a name: ").strip()
    age = input("How old are you? ").strip()
    while not age.isdigit():
        age = input("Please enter your age as a number: ").strip()
    user["name"] = name
    user["age"] = int(age)
    print(f"Nice to meet you, {user['name']}! Let's make your room fire. üî•")

def GetResponse(text):
    key = text.lower().strip()
    if key in responses:
        return responses[key]
    for k in responses:
        if k in key:
            return responses[k]
    return "Hmm... idk what to say ü§î"

def EndChat():
    print(responses.get("bye", "Bye!"))

def ShowStatus():
    print("\n--- STATUS ---")
    print(f"Energy: {state['energy']}/100  Mood: {state['mood']}  Points: {state['points']}")
    print(f"Room Level: {state['room_level']}  Decor items: {len(state['decor_items'])}")
    print("----------------\n")

def ShopMenu():
    print("\nüõí ROOM DECOR SHOP üõãÔ∏è")
    for name, info in shop_items.items():
        owned = " (owned)" if name in state["decor_items"] else f" ‚Äî {info['cost']} pts"
        print(f"- {name.title()}{owned}: {info['desc']}")

    while True:
        choice = input("Buy what? ").strip().lower()
        if choice == "exit":
            return "Leaving shop."
        if choice not in shop_items:
            print("That item isn't here. Try again.")
            continue
        if choice in state["decor_items"]:
            print("You already own that item.")
            continue
        
        cost = shop_items[choice]["cost"]
        if state["points"] < cost:
            print("Not enough points.")
            return "Shop closed."

        state["points"] -= cost
        state["decor_items"].append(choice)
        state["room_level"] += 1
        state["mood"] = "happy"

        if state["room_level"] % 3 == 0:
            old = state["energy"]
            state["energy"] = min(100, state["energy"] + 10)
            print(f"Room upgrade bonus: +{state['energy'] - old} energy!")

        return f"Purchased {choice.title()} for {cost} pts."

def UpdateStats(action):
    a = action.lower().strip()
    if a == "study":
        state["energy"] = max(0, state["energy"] - 20)
        state["points"] += 10
        state["mood"] = "productive"
        return "Studied: -20 energy, +10 points. üìö"

    if a == "eat":
        old = state["energy"]
        state["energy"] = min(100, state["energy"] + 15)
        state["mood"] = "satisfied"
        return f"Ate: +{state['energy'] - old} energy. üçî"

    if a in ("play", "play games", "game"):
        state["energy"] = max(0, state["energy"] - 10)
        state["points"] += 5
        state["mood"] = "excited"
        return "Played games: -10 energy, +5 points. üéÆ"

    if a == "sleep":
        state["energy"] = 100
        state["mood"] = "rested"
        return "Slept: energy ‚Üí 100. üò¥"

    if a == "shop":
        return ShopMenu()

    return "Unknown action."

def EndGame():
    print("\n=== GAME SUMMARY ===")
    ShowStatus()
    if state["decor_items"]:
        print("Your decor:")
        for item in state["decor_items"]:
            print(f"- {item.title()}")
    else:
        print("No decor yet.")
    print("Thanks for playing! üíñ")

def activity_menu():
    while True:
        ShowStatus()
        choice = input("Pick (study/eat/play/sleep/shop/quit): ").strip().lower()
        if choice in ("quit", "exit"):
            break
        print(UpdateStats(choice))
        time.sleep(0.5)

def conversation():
    print("Chat with your AI bestie.")
    while True:
        text = input("> ").strip()
        if not text:
            continue
        l = text.lower()
        if any(w in l for w in ("play", "game")):
            ans = input("Wanna play? (yes/no) ").strip().lower()
            if ans.startswith("y"):
                activity_menu()
                continue
        if l in ("bye", "goodbye", "quit", "exit"):
            EndChat()
            EndGame()
            break
        print(GetResponse(text))

if __name__ == "__main__":
    GreetUser()
    conversation()
```
