```python
import time

user = {}

responses = {
    "hi": "Hey bestie! üëã",
    "hello": "Hey bestie! üëã",
    "hey": "Hey bestie! üëã",
    "how are you": "Just vibing, you? üòé",
    "bye": "Aww, see you soon üíñ",
}

state = {
    "energy": 80,
    "mood": "neutral",
    "points": 0,
    "room_level": 1,
    "decor_items": [],
}

shop_items = {
    "poster": {"cost": 10, "desc": "Cute artistic poster ‚Äî +mood"},
    "plants": {"cost": 15, "desc": "Aesthetic plants ‚Äî chill vibes"},
    "lamp": {"cost": 20, "desc": "Cozy lamp ‚Äî better energy"},
    "beanbag": {"cost": 25, "desc": "Comfy beanbag ‚Äî relax mode"},
    "rgb lights": {"cost": 35, "desc": "RGB lights ‚Äî cool vibe"},
}

def GreetUser():
    print("‚ú® Welcome to GenZ Life Simulator ‚Äî Room Edition ‚ú®")
    name = input("What's your name? ").strip()
    while not name:
        name = input("Please type a valid name: ").strip()

    while True:
        age = input("How old are you? ").strip()
        if age.isdigit():
            break
        print("Please enter a number.")

    user["name"] = name
    user["age"] = int(age)
    print(f"Nice to meet you, {name}! üî•")

def GetResponse(text):
    msg = text.lower().strip()
    for key in responses:
        if key in msg:
            return responses[key]
    return "Hmm... idk what to say ü§î"

def EndChat():
    print(responses["bye"])

def ShowStatus():
    print("\n--- STATUS ---")
    print(f"Energy: {state['energy']}/100")
    print(f"Mood: {state['mood']}")
    print(f"Points: {state['points']}")
    print(f"Room Level: {state['room_level']}")
    print(f"Decor items: {len(state['decor_items'])}")
    print("----------------\n")

def ShopMenu():
    print("\nüõí ROOM DECOR SHOP üõãÔ∏è")
    for item, info in shop_items.items():
        owned = "(owned)" if item in state["decor_items"] else f"- {info['cost']} pts"
        print(f"{item.title():12} {owned:12} ‚Üí {info['desc']}")

    while True:
        choice = input("Buy what (or 'exit')? ").strip().lower()
        if choice == "exit":
            return "Leaving shop."
        if choice not in shop_items:
            print("Invalid item.")
            continue
        if choice in state["decor_items"]:
            print("Already owned.")
            continue
        
        cost = shop_items[choice]["cost"]
        if state["points"] < cost:
            print("Not enough points.")
            return "Shop closed."

        state["points"] -= cost
        state["decor_items"].append(choice)
        state["room_level"] += 1
        state["mood"] = "happy"

        if state["room_level"] % 3 == 0:
            state["energy"] = min(100, state["energy"] + 10)
            print("Room bonus: +10 energy!")

        return f"Purchased {choice.title()}!"

def UpdateStats(a):
    if a == "study":
        state["energy"] = max(0, state["energy"] - 20)
        state["points"] += 10
        state["mood"] = "productive"
        return "Studied: -20 energy, +10 points."

    if a == "eat":
        gain = min(100 - state["energy"], 15)
        state["energy"] += gain
        state["mood"] = "satisfied"
        return f"Ate: +{gain} energy."

    if a in ("play", "play games", "game"):
        state["energy"] = max(0, state["energy"] - 10)
        state["points"] += 5
        state["mood"] = "excited"
        return "Played: -10 energy, +5 points."

    if a == "sleep":
        state["energy"] = 100
        state["mood"] = "rested"
        return "Slept: energy restored."

    if a == "shop":
        return ShopMenu()

    return "Unknown action."

def EndGame():
    print("\n=== GAME SUMMARY ===")
    ShowStatus()
    if state["decor_items"]:
        print("Decor Purchased:")
        for d in state["decor_items"]:
            print("-", d.title())
    else:
        print("No decor purchased.")
    print("Game over üíñ")

def activity_menu():
    while True:
        ShowStatus()
        cmd = input("(study/eat/play/sleep/shop/quit): ").strip().lower()
        if cmd in ("quit", "exit"):
            break
        print(UpdateStats(cmd))
        time.sleep(0.5)

def conversation():
    print("Chat with your AI bestie!")
    while True:
        msg = input("> ").strip().lower()
        if not msg:
            continue

        if "play" in msg:
            if input("Play? (yes/no): ").startswith("y"):
                activity_menu()
                continue

        if msg in ("bye", "exit", "quit"):
            EndChat()
            EndGame()
            break

        print(GetResponse(msg))

if __name__ == "__main__":
    GreetUser()
    conversation()
```
